```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
from chardet import detect 

from PIL import Image
import requests
from io import BytesIO

import warnings
warnings.simplefilter('ignore')
```

## 文字列のバイナリ化

```python
text = "いろはにほへとちりぬるを"

# バイナリ変換
text_bin = text.encode()
print(text_bin)

# encode の確認
print(detect(text_bin))

# バイナリ -> 文字列
print(text_bin.decode("utf-8"))
```

## 画像のバイナリ変換

```python
url = "https://raw.githubusercontent.com/GitHub30/sketch2code-cli/master/samples/sample1.jpg"

response = requests.get(url)
img = Image.open(BytesIO(response.content))
plt.imshow(img)
plt.title("元画像")
plt.show()

# numpyのarray化
img_np = np.array(img)

# numpy の型確認
print(img_np.dtype)

# バイナリ変換
img_bin = img_np.tobytes()
print(img_bin[:50])

# エンコード確認
print(detect(img_bin))

# バイナリ -> numpy 変換
plt.imshow(np.frombuffer(img_bin, dtype = np.uint8).reshape(img_np.shape))
plt.title("バイナリ->numpy array")
plt.show()
```
